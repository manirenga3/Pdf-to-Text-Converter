```python
import PyPDF2
file_name ="ebook-sql_serve"
pdfFileObject = open(file_name+ ".pdf", 'rb')
pdfReader = PyPDF2.PdfFileReader(pdfFileObject)
print(" No. Of Pages :", pdfReader. numPages)
for i in range(0,pdfReader.numPages):
    pageObject = pdfReader.getPage(i)
    print("Processing page {} of {}:".format(i,pdfReader. numPages))
    with open(file_name+".txt","a+",encoding='utf-8') as file:
        file.write(pageObject.extractText())
        file.write("\n"*2)
        file.write("*"*60)
        file.write("\n"*2)
        

#pageObject = pdfReader. getPage(0)

pdfFileObject. close()
            
```
